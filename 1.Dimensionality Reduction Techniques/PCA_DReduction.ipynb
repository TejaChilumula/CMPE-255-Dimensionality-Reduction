# üöÄ Exploring Dimensionality Reduction with PCA üß†

This notebook explores the application of Principal Component Analysis (PCA) on the Wine dataset to visualize and understand the impact of dimensionality reduction.

## Dataset üç∑üìä

We're using the Wine dataset, which contains measurements of various chemical constituents present in wines.

## Code Overview üñ•Ô∏èüîç

### Visualization Before Reduction
- Display original data representations in the feature space.

### PCA Dimensionality Reduction
- Implement PCA to transform the data into a lower-dimensional space.

### Visualization After Reduction
- Showcase the reduced data representations using PCA.

## Visuals üìàüîç

The notebook contains pairs of visualizations showcasing the dataset's transformation before and after PCA. Each pair demonstrates the original data's distribution and the reduced representation in a lower-dimensional space.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA

# Loading the dataset
wine = load_wine()
data = wine.data
target = wine.target


fig, axes = plt.subplots(5, 2, figsize=(12, 20))

for i in range(5):
    # Displaying the original data from the dataset - wine (features 0 and 1)
    axes[i, 0].scatter(data[:, 0], data[:, 1], c=target, cmap='viridis', edgecolor='none', alpha=0.8)
    axes[i, 0].set_title(f'Original {i+1}')
    axes[i, 0].set_xlabel('Feature 0')
    axes[i, 0].set_ylabel('Feature 1')

    # Applying PCA for dimensionality Reduction
    pca = PCA(n_components=2)
    reduced_data = pca.fit_transform(data)

    # Visualize the reduced data after dimensionality reduction
    axes[i, 1].scatter(reduced_data[:, 0], reduced_data[:, 1], c=target, cmap='viridis', edgecolor='none', alpha=0.8)
    axes[i, 1].set_title(f'Reduced {i+1}')
    axes[i, 1].set_xlabel('Principal Component 1')
    axes[i, 1].set_ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```

## Conclusion üåüüßê
